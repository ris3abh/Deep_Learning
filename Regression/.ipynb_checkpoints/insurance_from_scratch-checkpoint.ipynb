```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('./insurance_data.csv')
df.head()
```

```python
def sigmoid(z):
    return 1 / (1 + np.exp(-z))
```

```python
coef = [9.097272 ,1.7341233 ]
intercept = -4.852654
```

```python
def prediction(age ,affordiblity):
    x = coef[0] * age/100
    y = coef[1] * affordiblity
    w_s = x + y + intercept 
    return sigmoid(w_s)
```

```python
print(prediction(47,1))
```

```python
print(prediction(22,1))
```

```python
def log_loss(y_true ,y_predicted):
    epsilon = 1e-15
    y_predicted_new = [max(i,epsilon) for i in y_predicted]
    y_predicted_new = [min(i,1-epsilon) for i in y_predicted_new]
    y_predicted_new = np.array(y_predicted_new)
    return -np.mean(y_true * np.log(y_predicted_new) + (1-y_true)*np.log(1-y_predicted_new))
```

```python
def sigmoid_numpy(X):
    return 1 / (1 + np.exp(-X)) 

##sigmoid_numpy(np.array([1,12,8]))
```

```python
def gradient_descent(age, affordibility ,y_true ,epochs):
    w1 ,w2 ,bias  = 1, 1, 0
    rate = 0.5
    n =len(age)
    for i in range(epochs):
        w_s = w1 * age + w2 * affordibility + bias 
        y_predicted = sigmoid_numpy(w_s)
        loss = log_loss(y_true ,y_predicted)
        w1d = (1/n)*np.dot(np.transpose(age),(y_predicted - y_true)) 
        w2d = (1/n)*np.dot(np.transpose(affordibility),(y_predicted - y_true))
        bias_d = np.mean(y_predicted - y_true)
        w1 = w1 - (rate*w1d)
        w2 = w2 - (rate*w2d)
        bias = bias - rate* bias_d
        print(f'Epochs:{i}, w1:{w1} ,w2: {w2} ,bias:{bias} ,loss:{loss}')
    return w1 ,w2 ,bias

```

```python
from sklearn.model_selection import train_test_split
X_train ,X_test ,Y_train ,Y_test = train_test_split(df[['age','affordibility']], df.bought_insurance ,test_size = 0.2 ,random_state = 25)
```

```python
x_train_scaled = X_train.copy()
x_test_scaled = X_test.copy()

x_train_scaled['age'] = x_train_scaled['age']/100
x_test_scaled['age'] = x_test_scaled['age']/100
```

```python
gradient_descent( x_train_scaled['age'] ,X_train['affordibility'] ,Y_train ,5000)
```
